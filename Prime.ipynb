# Collatz Conjecture:

We will get 1 eventually


```python
m=9
for i in range(100):
    print(f"m_{i}=",m)
    if m==1:
        break
    m=f(m)

    
```

```python
x=[]
y=[]
for m in range(2000):
    if m==0:
        continue
    x.append(m)
    print("m=",m,end = ' ')
    iterator=1
    while m!=1:
        #print(f"m_{i}=",m)
        if m==1:
            break
        iterator+=1
        m=f(m)
    y.append(iterator)
    print(" OK! Solved with ",iterator)
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15, 10))
plt.plot(x,y)
plt.ylabel('some numbers')
plt.show()
```
